```python
#For loop iterates through all the columns
for i in df.columns:
    #if the data type of the column is int or float
    if df[i].dtypes in ['int64','float64']:
        #we find mean and standard deviation
        mean=np.mean(df[i])
        std=np.std(df[i])
        #we substitute mean and standard deviation in z score formule
        df[i+' Standard'] = (df[i]-mean)/std
```

Standardization or z score normalization is a step used in preprocessing data, it scales the features in the dataset on a similar scale. By doing so we improve perpormance of machine learning algorithm.

To Standardize the data we use z score, Formula for z score is x -mean of x/ Standard deviation of x

The Standarize value has mean of 0 and Standard Deviation of 1

Conversion of Categorical Data into Dummy Variables

    Objective: To transform categorical variables into a format that can be provided to ML algorithms.
    Steps:
        Discuss the need for converting categorical data into dummy variables (one-hot encoding).
        Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
        Display a portion of the transformed dataset.

For better performance of machine learning algorithm its is recomended to give numerical values, so we have to convert categorical values to numerical value.

The process of converting categorical value to numerical value is called labelling.

One Hot Encoder is used for labeling nominal data.

One Hot Encoder convert the categorical data into 0's and 1's.

There are three steps for One Hot Encoder,
1) Pass data to One Hot Encoder, it will return an 2d-array with zero and one. Zero if data is absent and one if data is present.

2) Then we convert that array into DataFrame.

3) Then join that DataFrame to a new_data and remove categorical data.

```python
#Seeing categorical data(Dtype = object)
df.info()
```

```python
#importing one hot encoder from preprocessing.
from sklearn.preprocessing import OneHotEncoder
```

```python
#Storing OnHotEncoder() in a variable
one_hot=OneHotEncoder()
```

```python
df.head()
```

```python
#Creating an array for categorical data
columns=[]
for i in df.columns:
    if(df[i].dtypes=='object'):
        #append the data into the array if data is categorical
        columns.append(i)

```

```python
#Creating a function for labeling the data
def labeling(i):
    #Processing the data and storing it in an variable
    new_df = df.join(pd.DataFrame(one_hot.fit_transform(df[[i]]).toarray()))
    return new_df #Returning the variable
```

```python
for col in columns:
    #Calling the function inside the loop
    new_df=labeling(col)
```

```python
for col in columns:
    #Droping the Categorical Data
    new_df.drop(str(col),axis=1,inplace=True)
```
